```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task1
data = pd.read_csv("D:/Aman/Cognifyz.csv")
print(data)
print(data.info())
data.describe()
```

```python
#Task2
print(data["gender"])
df = pd.DataFrame(data)
gender_counts = df['gender'].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.2f%%', startangle=100, )
plt.title('Gender Distribution in Dataset')
plt.show()
```

```python
#Task3
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
numerical_columns

statistics = pd.DataFrame({
    'Mean': df[numerical_columns].mean(),
    'Median': df[numerical_columns].median(),
    'Standard Deviation': df[numerical_columns].std()
})
statistics
```

```python
#Task4
avenue_counts = df['Avenue'].value_counts()

most_common_avenue = avenue_counts.idxmax()
most_common_avenue_count = avenue_counts.max()

most_common_avenue, most_common_avenue_count
```

```python
#Task5
reason_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

for col in reason_columns:
    print(f"--- Analysis for column: '{col}' ---")
    print(df[col].value_counts())
    print("\n")

reason_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

all_reasons = pd.concat([df[col] for col in reason_columns]).dropna()

theme_mapping = {
    'Capital Appreciation': 'Returns/Growth',
    'Dividend': 'Returns/Growth',
    'Better Returns': 'Returns/Growth',
    'High Interest Rates': 'Returns/Growth',
    'Fixed Returns': 'Returns/Growth',
    'Assured Returns': 'Returns/Growth',
    'Safe Investment': 'Safety/Security',
    'Risk Free': 'Safety/Security',
    'Fund Diversification': 'Diversification',
    'Tax Benefits': 'Tax Benefits',
    'Tax Incentives': 'Tax Benefits',
    'Liquidity': 'Liquidity'
}

investment_themes = all_reasons.map(theme_mapping)
theme_counts = investment_themes.value_counts()
print("Common Investment Themes and their Frequencies:")
print(theme_counts)
```
